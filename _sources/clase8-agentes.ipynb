# 14. Agentes en finanzas

## QuantAgent

Un sofisticado sistema de análisis de trading multiagente que combina indicadores técnicos, reconocimiento de patrones y análisis de tendencias mediante LangChain y LangGraph. El sistema ofrece una interfaz web y acceso programático para un análisis exhaustivo del mercado.

[https://github.com/Y-Research-SBU/QuantAgent](https://github.com/Y-Research-SBU/QuantAgent)

## TradingAgents

TradingAgents es un flujo de trading multiagente que refleja la dinámica de las empresas de trading del mundo real. Mediante la implementación de agentes especializados con tecnología LLM, desde analistas fundamentales, expertos en sentimiento y analistas técnicos hasta operadores y equipos de gestión de riesgos, la plataforma evalúa colaborativamente las condiciones del mercado y fundamenta las decisiones de trading. Además, estos agentes participan en debates dinámicos para determinar la estrategia óptima.

[https://github.com/TauricResearch/TradingAgents](https://github.com/TauricResearch/TradingAgents)
